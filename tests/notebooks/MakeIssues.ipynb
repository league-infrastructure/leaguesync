```python
mod_dir = Path('/Users/eric/proj/league-projects/curriculum/league-java-modules')

import re  # Importing the regular expression library

# Given list of names
names = list(mod_dir.glob('Level*'))
names = [e.name for e in names]
# Function to extract first and second numbers from each entry
def extract_numbers(names):
    extracted = []
    for name in names:
        name = str(name)
        # Split the string by non-digit characters and filter out empty strings
        numbers = [int(num) for num in filter(None, re.split("\D", name))]
        # Append the first two numbers found to the extracted list
        extracted.append(numbers[:2])
    return extracted

# Extract numbers and print the results
extracted_numbers = extract_numbers(names)

import pandas as pd
numbers = list(sorted(extracted_numbers))
t = pd.DataFrame(numbers, columns = ['level','module'])
t['name'] = names
t

```
